```python
import requests
import urllib.request
from bs4 import BeautifulSoup 
```

```python
r=requests.get("https://www.yellowpages.com/los-angeles-ca/coffee-shops")
r.content

```

```python
soup=BeautifulSoup(r.content)
soup
```

```python
soup.findAll("a")
```

```python
for link in soup.findAll("a"):
    print(link.text,link.get('href'))
```

```python
data=soup.findAll('div',{'class':'info'})
for item in data:
    print(item.contents)
```

```python
shop_name=[]
name=soup.findAll("a",{"class":"business-name"})
for item in name:
    shop_name.append(item.text)
shop_name
```

```python
address=[]
ad=soup.findAll("div",{"class":"street-address"})
for adr in ad: 
    address.append(adr.text)
address
```

```python
phone=[]
ph=soup.findAll("div",{"class":"phones phone primary"})
for ph in p: 
    phone.append(ph.text)
phone 
```

```python
locality=[]
l=soup.findAll("div",{"class":"locality"})
for loc in l: 
    locality.append(loc.text)
locality
```

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.DataFrame({"Coffee Shop":shop_name,"Street address":address,'Locality':locality,'Contact':phone})
df.to_csv('coffee shops in LA.csv',index=False)
df
```
